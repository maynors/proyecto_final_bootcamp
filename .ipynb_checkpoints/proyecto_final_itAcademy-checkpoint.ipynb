## Agregando nuevas columnas

```python
# Recordemos que el activo y Fondos propios son una "fotografía" realizada en un día en específico,en este caso el 31-diciembre de cada año.
# Para tener un valor más acorde a nuestro estudio se realiza el promedio de los dos años.
df['Activo2018_2017']= (df.Activo2018+df.Activo2017)*0.5
df['FP2018_2017']= (df.FondosPropios2018+df.FondosPropios2017)*0.5
df.head(2)
```

### Agregar columnas con los indicadores(KPI's) a analizar

```python
df['MargenVentas']= (df.EBIT2018 / df.CifraVentas2018)
df['RotacionActivo']= (df.CifraVentas2018 / df.Activo2018_2017)
df['ApalancamientoFinanciero']= ((df.Activo2018_2017 / df.FP2018_2017)*(df.Resultado_ord_antes_impuestos_BAI2018 / df.EBIT2018))
df['EfectoFiscal']= (df.ResultadoEjercicio2018 / df.Resultado_ord_antes_impuestos_BAI2018)
df.head(2)
```

```python
# Es interesante convertir la columna del Resultado Ejercicio en una variable dummy, si es >= 0 tendra el valor de 1, y si es < 0 tendra el valor de 0.
# De esta forma será más fácil analizar las empresas que tuvieran un resultado del ejercicio positivo o negativo.
df['BN2018']= df['ResultadoEjercicio2018']
df['BN2018'] = df['BN2018'].apply(lambda x: 1 if x >= 0 else 0)
```

```python
df.columns.tolist()
```

```python
# Eliminamos las columnas de Activo y FondosPropios, ya tenemos dos columnas con sus promedios
df= df.drop(['Activo2018','Activo2017','FondosPropios2018','FondosPropios2017'],axis=1)
```

## Análisis de outlieres para el indicador: Efecto fiscal

El efecto fiscal esta calculado de dos valores del estado de perdidas y ganancias: Beneficio Neto & BAI.  
El beneficio neto es el resultado de restar el BAI - impuestos. Por lo cual siempre debería ser menor a 1 y mayor a 0. El impuesto debería aplicarse si se tiene un BAI positivo. 

```python
#Gráfica del Efecto Fiscal
sns.boxplot(y='EfectoFiscal', data=df)

# Mostrar el gráfico
plt.title('BoxPlot del efecto Fiscal')
plt.show()
```

> El boxplot tiene datos totalmente alejados del concepto (<1) y visualmente tampoco es útil. Procedemos a eliminar todos lo valores mayor a 1 y menor a 0.


```python
#Eliminar los valores atípicos del Efecto Fiscal
df = df[(df['EfectoFiscal'] >= 0) & (df['EfectoFiscal'] <= 1)]
```

```python
#Gráfica del Efecto Fiscal
sns.boxplot(y='EfectoFiscal', data=df)

# Mostrar el gráfico
plt.title('BoxPlot del efecto Fiscal')
plt.show()
```

> Ahora se se observan un Boxplot más acorde a lo que pasa en la realidad. El boxplot muestra el ratio financierodel Efecto Fiscal, si restamos 1 - EfectoFiscal= % de impuesto a pagar

> La mayoría de empresas pagan entre 25% o 0% de impuestos. El 0 de impuestos es aquellas empresas que no tuvieron beneficios y por lo tanto no se les cobra impuestos.

> Es importante destacar que el modelo de mapas auto organizativos tiene la capacidad de trabajar con outliers y valores "atípicos", lo que hace el modelo con estos valores es que los agrupa en pequeños clusters y el analista puede observar que son valores erroneos o extraños. En nuestro caso se realiza este procesado para que el resultado sea más limpio.

```python
# Eliminamos los valores atípicos del Apalancamiento Financiero
df= df[(df['ApalancamientoFinanciero'] >= -20) & (df['ApalancamientoFinanciero'] <= 20)]
```

# Aplicación del modelo no supervisado: mapas auto-organizados.

El algoritmo funciona basicamente de la siguiente manera:
    
1) Se ingresan un número n de registros con las variables que el analista decida. En este caso ingresaremos 5 variables
* Margen de Ventas
* Rotacion de Activo
* Apalancamiento Finaciero
* Efecto Fiscal
* Beneficio Neto (1 si es positivo y 0 si es negativo)

2) Se elige el número de neuronas ganadoras (clusters) con base al número de registros. Existe una formula:

* 5*√1820 =213 . Dónde 1820 es el número de registros. Ahora tendremos que elegir dos números que multplicados me de cercano a 213. 
* Elegimos 15*14= 210. Existiran 210 pequeños clusters.


3) Se observan los resultados de forma gráfica y se analizan. Esta es de las partes más importantes ya que es donde obtenemos las respuestas y conclusiones a dudas.

A continuación una imagen que resumen los puntos descritos anteriormente:


> Cada hexagono representa un grupo de registros con caracteristicas similares. Una de las ventajas es que se puede observar como se comporta cada variable en cada una de los hexagonos. Por ejemplo:


```python
## Entrenamiento del modelo
```

```python
# Importamos la librería del modelo
import simpsom as sps

# Federico Comitani. (2022). SimpSOM (v2.0.2). Zenodo. https://zenodo.org/record/7187332
```

```python
# Elegimos el número de neuroas ganadoras(pequeños clusters). En nuestro caso 15 *14 = 210
som_width = 15
som_length = 14
```

```python
#Alimentamos la red
simp_net = sps.SOMNet(som_width, som_length, df[['MargenVentas', 'RotacionActivo',
       'ApalancamientoFinanciero', 'EfectoFiscal', 'BN2018']].to_numpy(), PBC=True)
simp_net.train()
```

# MAPA AUTO ORGANIZADO:  MARGEN DE VENTAS

```python
simp_net.nodes_graph(colnum=0)
```

> Se observa que la mayoría de empresas tiene un margen de ventas positivo, sin embargo es un margen muy bajo. Era de esperar al tratarse del sector panadero, donde generalmente producen productos "baratos" pero de alta rotación.
